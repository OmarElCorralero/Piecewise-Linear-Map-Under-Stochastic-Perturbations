```python
import numpy as np
import random
import matplotlib.pyplot as plt
import time

#Iterations
NN = 32000

#Parameters
a = 1/2 
b = 1
noise_level = 0.001
e = 1 #If you want the deterministic version replace 'e' by 0

Xf = np.zeros(NN)
Yf = np.zeros(NN)
epsif = np.zeros((2, NN))

# Initial Condition
Xf[0] = 2.1
Yf[0] = 2.1

for i in range(NN):
    random.seed(int(time.time()) + i)
    epsif[0, i] = random.gauss(0, noise_level)
    random.seed(i + NN)
    epsif[1, i] = random.gauss(0, noise_level)

for i in range(NN - 1):
    Xf[i + 1] = 1 - b * Yf[i] + 2 * a * abs(Xf[i]) + e * epsif[0, i]
    Yf[i + 1] = Xf[i] + e * epsif[1, i]

#To see the evolution every 8000 points
Xf2 = Xf[:NN//4]
Yf2 = Yf[:NN//4]
Xf4 = Xf[NN//4:NN//2]
Yf4 = Yf[NN//4:NN//2]
Xf6 = Xf[NN//2:3*NN//4]
Yf6 = Yf[NN//2:3*NN//4]
Xf8 = Xf[3*NN//4:]
Yf8 = Yf[3*NN//4:]
```

```python
#Phase Portrait

plt.scatter(Xf2, Yf2, s=2, c='blue')
plt.scatter(Xf4, Yf4, s=2, c='black')
plt.scatter(Xf6, Yf6, s=2, c='orange')
plt.scatter(Xf8, Yf8, s=2, c='purple')

# Initial condition
plt.scatter(Xf2[0], Yf2[0], s=10, c='green', label='Initial Condition')

plt.xlabel('x[n]')
plt.ylabel('y[n]')
plt.title('a=1/2, b=1 and Noise Level=0.001')
plt.legend()
plt.grid(True)


plt.show()
```
