```python
import numpy as np
import random
import matplotlib.pyplot as plt
import time
import pandas as pd
```

```python
#Iterations
NN = 32000

#Parameters

a = 1/2
b = 1
noise_level = 0.01
e = 1 #If you want the stochastic version replace 'e=0' by 'e=1'

X = np.zeros(NN)
Y = np.zeros(NN)
epsi = np.zeros((2, NN))

# Initial Condition
X[0] = 2.1
Y[0] = 2.1

for i in range(NN):
    random.seed(int(time.time()) + i)
    epsi[0, i] = random.gauss(0, noise_level)
    random.seed(i + NN)
    epsi[1, i] = random.gauss(0, noise_level)

for i in range(NN - 1):
    X[i + 1] = 1 - b * Y[i] + 2 * a * abs(X[i]) + e * epsi[0, i]
    Y[i + 1] = X[i] + e * epsi[1, i]

#To see the evolution every 8000 points
X2 = X[:NN//4]
Y2 = Y[:NN//4]
X4 = X[NN//4:NN//2]
Y4 = Y[NN//4:NN//2]
X6 = X[NN//2:3*NN//4]
Y6 = Y[NN//2:3*NN//4]
X8 = X[3*NN//4:]
Y8 = Y[3*NN//4:]
```

```python
#Phase Portrait

plt.scatter(X2, Y2, s=0.5, c='blue')
plt.scatter(X4, Y4, s=0.5, c='black')
plt.scatter(X6, Y6, s=0.5, c='orange')
plt.scatter(X8, Y8, s=0.5, c='purple')

# Initial condition
plt.scatter(X2[0], Y2[0], s=10, c='green', label='Initial Condition')

plt.xlabel('x[n]')
plt.ylabel('y[n]')
if e == 1:
    plt.title(f'a={a}, b={b} and Noise Level={noise_level}')
else:
    plt.title(f'a={a}, b={b}')
plt.legend()
plt.grid(True)


plt.show()
```
